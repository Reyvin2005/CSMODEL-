# Dataset Description


## Description of Dataset

* The dataset that was chosen is the Survey on Overseas Filipino Workers

  The purpose of the survey was to collect data about the number of overseas workers in the Philippines, their socioeconomic status, and other relevant details about foreign workers who were employed between April and September of 2015.

## Collection of Data
The data collection for the 2015 Survey on Overseas Filipinos (SOF) involved face-to-face interviews, leveraging the 2003 Master Sample, which is designed for household surveys. The survey targeted Filipino workers who had worked abroad between April and September 2015, focusing on their socio-economic characteristics and remittances. Using the Labor Force Survey as a sampling base, it provided national and regional estimates. Surpervised by a Supervision Team consisting of Regional Directors (RDs), Provincial Statistical Officers (PSOs) and a three-stage sampling process. The survey used a stratified, multistage sampling design. It focuses on 17 administrative regions in the Philippines, using area samples (Primary Sampling Units) selected with probability proportional to size. Both were done to ensure data quality, but the sample's limitation to private households excludes institutional populations.

The two questionnares that were used are:

SOF Form 1: Two-page questionnaire to gather data on the socio-economic characteristics of overseas Filipinos (OFs) and their remittances. Each overseas member received a separate form if multiple members were abroad during the reference period.

SOF Form 2: Household control form, used to determine if households have members who are overseas.

**Data Collection**

The data collection method of the 2015 Survey on Overseas Filipinos (SOF), which involved face-to-face interviews and was linked to the Labor Force Survey, impacts the accuracy and representativeness of the insights. Enumerators selected households from the sample, using questionnaires to gather data on overseas Filipino workers' socio-economic characteristics and remittances. The survey excluded institutional populations (e.g. dormitories, hospitals). While the sampling design ensures regional representation, it excludes institutional populations and relies on self-reported data, which can introduce biases.

For RQ8_HGRADE

```python
def convert_education_level(grade):
    # Extract the first digit as an integer
    first_digit = int(str(grade)[0])

    # Classify based on the first digit
    if first_digit == 0:
        return 'No Grade Completed'
    elif first_digit == 1:
        return 'Elementary Undergraduate'
    elif first_digit == 2:
        return 'Elementary Graduate'
    elif first_digit == 3:
        return 'High School Undergraduate'
    elif first_digit == 4:
        return 'High School Graduate'
    elif first_digit == 5:
        return 'Post Secondary Non-Tertiary Technical-Vocational Undergraduate'
    elif first_digit == 6:
        return 'College Undergraduate'
    else:
        return 'Unknown'  # Handle unexpected values

df['RQ8_HGRADE'] = df['RQ8_HGRADE'].apply(convert_education_level)
```

## Clustering

For clustering, we have decided to use all of the features instead of pair-wise as it can provide more comprehensive insights and lead to more accurate results. Using pair-wise may simplify our analysis but we might miss important relationship in the process.

Let's combine the one-hot encoded and standardized values to a single dataframe first.


```python
categorical_cols = ['RQ7_MSTAT', 'RQ8_HGRADE', 'RQ9_USOCC', 'RQ10_REASON', 'RQ11_BASE', 'RQ12_CTRY', 'RQ14_INDWORK', 'RQ15_OCCUP1']
df_encoded = pd.get_dummies(df, columns=categorical_cols, dtype=int)


scaler = StandardScaler()
numerical_cols = ['RQ4_AGE', 'RQ5_TMSLEFT', 'RQ13_STAY', 'RQ16_NOMONTH', 'RQ22_CASHAMT']
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])


df_encoded.drop(columns=['RQ4_AGE', 'RQ5_TMSLEFT', 'RQ13_STAY', 'RQ16_NOMONTH', 'RQ22_CASHAMT'], inplace=True)
df_encoded.info()
```

We are going to use K-means as way to cluster and let's take a look at each cluster information.

First we must determine what K-value to use. We can do this by utilizing the elbow method.

```python
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(df_encoded)
  wcss.append(kmeans.inertia_)  # Inertia is the WCSS

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

We can see that the curve starts to inflect at K=4 so we will set our K to that

```python
kmeans = KMeans(n_clusters=4, random_state=42)  # Adjust n_clusters as needed
df_encoded['Cluster'] = kmeans.fit_predict(df_encoded)

cluster_summary = df_encoded.groupby('Cluster').mean()
print(cluster_summary)
```

Now we look at how many observations or rows are in each cluster. With Cluster 2 being the highest amount.

```python
print("\nCluster Distribution:")
print(df_encoded['Cluster'].value_counts())
```

```python
# Apply PCA to reduce the data to 2D for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_encoded.drop('Cluster', axis=1))

plt.figure(figsize=(18, 16))
plt.scatter(pca_components[:, 0], pca_components[:, 1], c=df_encoded['Cluster'], cmap="magma")
plt.title("PCA of Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')

#plt.xlim([-5, 10])  # Example limits, adjust as needed
#plt.ylim([-5, 10])  # Example limits, adjust as needed
plt.show()
```

```python
loadings = pca.components_
loadings_df = pd.DataFrame(loadings.T, columns=['PC1', 'PC2'], index=df_encoded.drop('Cluster', axis=1).columns)
loadings_df["PC1"].sort_values(ascending=False)
loadings_df["PC2"].sort_values(ascending=False)
```
